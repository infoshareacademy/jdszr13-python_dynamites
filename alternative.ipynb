```python
UE_country = ["Austria", "Belgium", "Bulgaria", "Croatia", "Cyprus", "Czech Republic", "Denmark", "Estonia",
              "Finland", "France", "Greece", "Spain", "Netherlands", "Ireland", "Lithuania", "Luxembourg",
              "Latvia", "Malta", "Germany", "Poland", "Portugal", "Romania", "Slovakia", "Slovenia", "Sweden", 
              "Hungary", "Italy", "United Kingdom"]
df_test_1 = df[
    (df["country_or_area"].isin(UE_country))
    & 
    (df["year"] >= 2009) 
    & 
    (df["category"] == "kerosene_type_jet_fuel")
    &
    (   
        (df["commodity_transaction"] == "Kerosene-type Jet Fuel - Consumption by domestic aviation")
        |
        (df["commodity_transaction"] == "Kerosene-type Jet Fuel - International aviation bunkers")
    )

]


Aviation_in_EU = df_test_1[["country_or_area", "commodity_transaction", "year", "unit", "quantity", "category"]]
Aviation_in_EU.reset_index(drop = True, inplace = True)
Aviation_in_EU
 
sum_jet_fuel_UE = Aviation_in_EU.groupby("country_or_area")["quantity"].sum().reset_index()
sum_jet_fuel_UE.columns = ["EU countries", "sum_jet_fuel"]
sum_jet_fuel_UE["sum_jet_fuel"] /= 1000
sum_jet_fuel_UE


```

```python
import matplotlib.pyplot as plt
sum_jet_fuel_UE.sort_values(["sum_jet_fuel"]).plot(kind = "bar", x = "EU countries", y = "sum_jet_fuel" , figsize=(12, 5))
plt.title("Kerosene-type Jet Fuel - Consumption by aviation" " \n" "(domestic and international flights 2009 - 2014)", fontsize = 18)
plt.xlabel("EU countries", fontsize = 15)
plt.ylabel("Burnt Jet Fuel [tones x 1000000]", fontsize = 15)
plt.show()
```

```python
World = ["United States", "China", "India", "Russian Federation"]
df_test_2 = df[
    (df["country_or_area"].isin(World))
    & 
    (df["year"] >= 2009) 
    & 
    (df["category"] == "kerosene_type_jet_fuel")
    &
    (   
        (df["commodity_transaction"] == "Kerosene-type Jet Fuel - Consumption by domestic aviation")
        |
        (df["commodity_transaction"] == "Kerosene-type Jet Fuel - International aviation bunkers")
    )

]

df_test_2

Aviation_in_World = df_test_2[["country_or_area", "commodity_transaction", "year", "unit", "quantity", "category"]]
Aviation_in_World.reset_index(drop = True, inplace = True)
Aviation_in_World
sum_jet_fuel_World = Aviation_in_World.groupby("country_or_area")["quantity"].sum().reset_index()
sum_jet_fuel_World.columns = ["World countries", "sum_jet_fuel"]
sum_jet_fuel_World["sum_jet_fuel"] /= 1000
sum_jet_fuel_World
```
