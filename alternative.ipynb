```python
#Kraje uni europejskiej, ilość paliwa lotniczego dla każdego państwa europejskiego (loty krajowe i międzynarodowe) 
UE_country = ["Austria", "Belgium", "Bulgaria", "Croatia", "Cyprus", "Czechia", "Denmark", "Estonia",
              "Finland", "France", "Greece", "Spain", "Netherlands", "Ireland", "Lithuania", "Luxembourg",
              "Latvia", "Malta", "Germany", "Poland", "Portugal", "Romania", "Slovakia", "Slovenia", "Sweden", 
              "Hungary", "United Kingdom", "Italy"]
df_filtr_1 = df[
    (df["country_or_area"].isin(UE_country))
    & 
    (df["year"] >= 2014) 
    & 
    (df["category"] == "kerosene_type_jet_fuel")
    &
    (   
        (df["commodity_transaction"] == "Kerosene-type Jet Fuel - Consumption by domestic aviation")
        |
        (df["commodity_transaction"] == "Kerosene-type Jet Fuel - International aviation bunkers")
    )

]

EU_population = [8.5, 11.2, 7.2, 4.3, 0.9, 10.5, 5.6, 1.3, 5.5, 66.0, 80.9, 10.9, 9.8, 4.6, 59.7, 2.0, 2.9, 0.5, 0.4, 16.8, 38.5, 10.4, 19.8, 5.4, 2.1, 46.7, 9.7, 64.1]
World_population = [1390, 1270, 143.7, 318.9, 506.2] 

Aviation_in_EU = df_filtr_1[["country_or_area", "commodity_transaction", "year", "unit", "quantity", "category"]]
Aviation_in_EU.reset_index(drop = True, inplace = True)
Aviation_in_EU
 
sum_jet_fuel_UE = Aviation_in_EU.groupby("country_or_area")["quantity"].sum().reset_index()
sum_jet_fuel_UE.columns = ["EU countries", "sum_jet_fuel"]
sum_jet_fuel_UE["sum_jet_fuel"] /= 1000
sum_jet_fuel_UE
```

```python
sum_jet_fuel_UE_population = sum_jet_fuel_UE.sort_values(by = ["EU countries"])
sum_jet_fuel_UE_population["UE_population"] = EU_population
sum_jet_fuel_UE_population["fuel_for_one_person_in_kg"] = (sum_jet_fuel_UE_population["sum_jet_fuel"] / sum_jet_fuel_UE_population["UE_population"])*1000
sum_jet_fuel_UE_population
```

```python
plt.figure(figsize=(12, 5))
sorted_sum_jet_fuel_UE = sum_jet_fuel_UE.sort_values(["sum_jet_fuel"])
sns.barplot(data = sorted_sum_jet_fuel_UE, x = "EU countries", y = "sum_jet_fuel")
plt.title("Kerosene-type Jet Fuel - Consumption by aviation" " \n" "(domestic and international flights in 2014)", fontsize = 18)
plt.xlabel("EU countries", fontsize = 15)
plt.ylabel("Burnt Jet Fuel [tones x 1000000]", fontsize = 15)
plt.xticks(rotation=90)
plt.show()
```
